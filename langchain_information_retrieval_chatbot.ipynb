```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma

embedding_model_name = "all-MiniLM-L6-v2"
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name)

vector_saving_directory = "./chromaaa_db"

vectorstore = Chroma(
    collection_name="pdf-documents",
    embedding_function=embeddings,
    persist_directory=vector_saving_directory,
)

if isinstance(texts[0], str):
    texts_content = texts
    metadatas = [{"source": "unknown"} for _ in texts_content]
else:
    texts_content = [doc.page_content for doc in texts]
    metadatas = [{"source": doc.metadata["source"]} for doc in texts]

vectorstore.add_texts(texts=texts_content, metadatas=metadatas)

```

```python


from langchain.prompts.prompt import PromptTemplate
custom_template = """Given the following conversation and a follow up question, rephrase the follow up question to be a standalone question.

Chat History:
{chat_history}

Follow Up Input:{question}
Standalone question:"""
CONDENSE_QUESTION_PROMPT = PromptTemplate.from_template(custom_template)
```

```python
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 5})
```

```python
# Checking LLM response..
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from transformers import pipeline

hf_model_name = "cmarkea/bloomz-560m-sft-chat"

hf_pipeline = pipeline(
    "text-generation",
    model=hf_model_name,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
    do_sample=True,
    max_new_tokens=150
)

llm = HuggingFacePipeline(pipeline=hf_pipeline)

PROMPT_TEMPLATE = """
Please answer the following question:
Question: {question}
Answer:"""

prompt = PromptTemplate(template=PROMPT_TEMPLATE, input_variables=["question"])

llm_chain = LLMChain(llm=llm, prompt=prompt)

question = "Describe AI?"
answer = llm_chain.run({"question": question})

print(f"Q: {question}")
print(f"A: {answer}")
```
